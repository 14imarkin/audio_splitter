Создадим датафрейм

```python
columns = ['Track_ID', 'Stem', 'SIR', 'SAR', 'SDR']
df = pd.DataFrame(data, columns=columns)
```

Подсчитаем среднее по стемам

```python
metrics = ['SIR', 'SAR', 'SDR']
result = df.groupby('Stem')[metrics].agg(['mean', 'std', 'min', 'median', 'max'])
```
